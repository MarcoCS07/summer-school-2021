# Introducción

* La ecuación de onda

\begin{equation}
\nabla^{2}u-\frac{\partial^{2}u}{\partial t^{2}}=0
\end{equation}

* Surge al describir fenómenos relativos a la propagación de ondas en un medio continuo. 
* Los estudios de ondas acústicas, ondas de gravedad, ondas electromagnéticas y vibraciones mecánicas están basados en esta ecuación.

* En una dimension

\begin{equation}
\frac{\partial^{2}u}{\partial x^{2}}-\frac{\partial^{2}u}{\partial t^{2}}=0
\end{equation}

* Resolvamos el problema de una cuerda vibrante.


# La cuerda vibrante

* La historia comienza más o menos a mitad del siglo XVIII.
* La vibración de la cuerda de un violín.
* Jean Le Rond d'Alembert fue el primero en describir un modelo matematico del problema.
  - Bernouilli
  - Dirichlet
  - Euler
  - Fourier

## La cuerda

*Medio continuo unidimensional, elástico y que no ofrece resistencia a la flexión.*

## Las vibraciones

* Supongamos ahora que nuestra cuerda se encuentra sobre el plano $Oxu$.
* Nuestra cuerda es sometida a vibraciones transversales
  - Contenidas en el plano
  - Respecto de su posición de equilibrio
  - La posición de equilibrio coincide con el eje $Ox$
* Convenientemente elegimos el sistema de referencia $Oxu$ 
* La cuerda de longitud $l$ se halla contenida en el intervalo $\left[0;l\right]$.

## Las condiciones

* Flexibilidad sin resistencia:
  - La tensión en cada punto $x$ y en cada instante $t$ es tangente a la cuerda.
  - El modulo de la tension tiene por módulo $T\left(x,t\right)$. 

* El desplazamiento vestical $u(x,t)$ de la cuerda en el punto $x$ al instante $t$
  - La longitud del segmento $\left[a,a+\epsilon\right]$ al deformase viene dada por

\begin{equation}
s=\intop_{a}^{a+\epsilon}\sqrt{1+\left(\frac{\partial u}{\partial x}\right)^2}dx
\end{equation}

* Las oscilaciones son "pequeñas"
  - Podemos despreciar el termino cuadratico debajo del radical
  - Directo: $s\approx\epsilon$
  
\begin{equation}
\left(\frac{\partial u}{\partial x}\right)^2\approx0
\end{equation}

* Ambos caminos implican que, según la *ley de Hooke*, el módulo de la tensión, que es proporcional al estiramiento, es constante: $\left|T(x,t)\right|=T_o$.

## La ecuación

* Los movimientos se realizan solo en la dirección $Ou$
* Las vibraciones son pequeñas
    - Consideramos las aproximaciones $\sin\theta\approx\theta\approx\tan\theta\approx\displaystyle\frac{\partial u}{\partial x}$
    
* $\displaystyle\frac{\partial^2 u}{\partial x^2}$ es la aceleración vertical del elemento $dx$

* La ecuacion del movimiento es

\begin{equation}
T_o\frac{\partial ^2u}{\partial x^2}+F(x,t)=\rho (x)\frac{\partial ^2u}{\partial t^2}
\end{equation}

- $\rho\left(x\right)$ es la densidad de la cuerda
- $F\left(x,t\right)$ es una fuerza externa

## El problema de Cauchy

* Consideremos que la cuerda esta fija por los extremos.
  - $u(0,t) = u(l,t) = 0 \quad , \quad t\in\mathbb{R}$
* La ausencia de fuerzas externas aparte de la gravedad
  - $F(x,t) = 0 \quad , \quad t\in\mathbb{R}$

* Juntamos todas las condiciones que hemos establecido hasta ahora

\begin{equation}
\left\{ \begin{array}{ccccc}
{\displaystyle T_{o}\frac{\partial^{2}u}{\partial x^{2}}} & = & {\displaystyle \rho(x)\frac{\partial^{2}u}{\partial t^{2}}}\\
{\displaystyle u\left(0,t\right)=u\left(l,t\right)} & = & {\displaystyle 0} & , & t\in\mathbb{R}\\
{\displaystyle u\left(x,0\right)} & = & {\displaystyle f\left(x\right)} & , & x\in\left[0,l\right]\\
{\displaystyle {\left.\frac{\partial^{2}u}{\partial t^{2}}\right|}_{t=0}} & = & {\displaystyle g\left(x\right)} & , & x\in\left[0,l\right]
\end{array}\right.
\end{equation}

* Sponiendo que ...
  - Densidad constante
  - $T_{o} / \rho(x) = c^2$

\begin{equation}
\left\{ \begin{array}{ccccc}
{\displaystyle \frac{\partial^{2}u}{\partial x^{2}}} - {\displaystyle \frac{1}{c^2}\frac{\partial^{2}u}{\partial t^{2}}}& = & 0\\
{\displaystyle u\left(0,t\right)=u\left(l,t\right)} & = & {\displaystyle 0} & , & t\in\mathbb{R}\\
{\displaystyle u\left(x,0\right)} & = & {\displaystyle f\left(x\right)} & , & x\in\left[0,l\right]\\
{\displaystyle {\left.\frac{\partial^{2}u}{\partial t^{2}}\right|}_{t=0}} & = & {\displaystyle g\left(x\right)} & , & x\in\left[0,l\right]
\end{array}\right.
\end{equation}

# La simulación

* En el caso analitico
  - Soluciones del tipo $u\left(x,t\right) = \Xi\left(x\right) \Phi\left(y\right)$
  - Separacion de variables

## La discretización

* Soluciones del tipo

\begin{equation}
u\left(x,t\right) = \Xi\left(x\right) \Phi\left(t\right)
\end{equation}

* Entonces

\begin{equation}
\frac{\partial^{2}u}{\partial x^{2}} = \frac{\partial^{2}\Xi \Phi}{\partial x^{2}} = \Phi\frac{\partial^{2}\Xi}{\partial x^{2}} = \Phi\frac{d^{2}\Xi}{d x^{2}}
\end{equation}

\begin{equation}
\frac{\partial^{2}u}{\partial t^{2}} = \frac{\partial^{2}\Xi \Phi}{\partial t^{2}} = \Xi\frac{\partial^{2}\Phi}{\partial t^{2}} = \Xi\frac{d^{2}\Phi}{d t^{2}}
\end{equation}

* Ya hemos visto las discretizacion de derivadas
  - La fija: Punto medio (Gracias Neisser :D)

\begin{equation}
\Phi\left(t\right)\left.\frac{d^{2}\Xi}{d x^{2}}\right|_{x=x_0} \approx 
    \Phi\left(t\right)\left[\frac{\Xi(x_0-h)-2\Xi(x_0)+\Xi(x_0+h)}{h^{2}}\right]
\end{equation}

\begin{equation}
\Xi\left(x\right)\left.\frac{d^{2}\Phi}{d t^{2}}\right|_{t=t_0} \approx 
    \Xi\left(x\right)\left[\frac{\Phi(t_0-\tau)-2\Phi(t_0)+\Phi(t_0+\tau)}{\tau^{2}}\right]
\end{equation}

* Volvemos a usar $u(x, t)$

\begin{equation}
\left.\frac{\partial^{2}u}{\partial x^{2}}\right|_{x=x_0} \approx 
    \frac{u(x_0-h,t)-2u(x_0,t)+u(x_0+h,t)}{h^{2}}
\end{equation}

\begin{equation}
\left.\frac{\partial^{2}u}{\partial t^{2}}\right|_{t=t_0} \approx 
    \frac{u(x, t_0-h)-2u(x, t_0)+u(x, t_0+h)}{\tau^{2}}
\end{equation}

* Ponemos todo en una ecuacion

\begin{equation}
\frac{u(x_0-h,t)-2u(x_0,t)+u(x_0+h,t)}{h^{2}} = \frac{1}{c^2} \frac{u(x, t_0-h)-2u(x, t_0)+u(x, t_0+h)}{\tau^{2}}
\end{equation}

* $\alpha^2 = \displaystyle \frac{c^2 \tau^{2}}{h^{2}}$ es el Numero de Courant.

* Despejamos $u(x, t+h)$

\begin{equation}
u(x, t_0+h) = 2\left( 1 - \alpha^2 \right) u(x_0,t) + \alpha^2 \left(u(x_0-h,t)+u(x_0+h,t)\right) - u(x, t_0-h)
\end{equation}

## La implementacion

```python
import numpy as np
```

```python
np_x = 101
np_t = 501

x_upper = 10
t_upper = 25

# h
delta_x = x_upper / (np_x - 1)
# tau
delta_t = t_upper / (np_t - 1)

# c^2 = 1
alpha_squared = pow(delta_t / delta_x, 2)
```

```python
alpha_squared
```

```python
def f(x):
    # return x * delta_x
    # return 2 * np.exp( -2 * np.power(x - 0.5 * x_upper, 2))
    # return (
    #     2 * np.exp( -2 * np.power(x - 0.25 * x_upper, 2)) -
    #     2 * np.exp( -2 * np.power(x - 0.75 * x_upper, 2))
    # )
    return 0
```

```python
def g(t):
    return 4 * np.exp( -2 * np.power(t - 0.03 * t_upper, 2))
```

```python
X = np.linspace(0, x_upper, np_x)
T = np.linspace(0, t_upper, np_t)

XT = np.zeros((np_x, np_t))
XT[:,0] = f(X)
XT[-1,:] = g(T)
```

```python
# Aplicamos euler para obtener el valore de u(x,1)
for x in range(1, np_x-1):
    XT[x, 1] = (1 - alpha_squared) * XT[x,0] + 0.5 * alpha_squared * (XT[x - 1, 0] + XT[x + 1, 0])
```

```python
for t in range(2, np_t):
    for x in range(1, np_x-1):
        XT[x, t] = 2 * (1 - alpha_squared) * XT[x, t - 1] + alpha_squared * (XT[x + 1, t - 1] + XT[x - 1, t - 1]) - XT[x, t - 2]
```
